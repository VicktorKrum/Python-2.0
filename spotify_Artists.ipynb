```python
# prompt: how to create a visual of top 5 songs that were streamed with a drill down for release year and month

import matplotlib.pyplot as plt
# Assuming your dataframe 'df1' has a 'release_year' and 'release_month' column
# If not, you'll need to extract or create these columns from your data

# Group by track_name and sum the streams
top_songs = df1.groupby('track_name')['streams'].sum().sort_values(ascending=False).head(5)

# Create a bar chart for top 5 songs
plt.figure(figsize=(10, 6))
plt.bar(top_songs.index, top_songs.values)
plt.xlabel('Song Name')
plt.ylabel('Total Streams')
plt.title('Top 5 Most Streamed Songs')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


```

```python
import plotly.express as px
```

```python


# Group by track_name and sum the streams for the top 5 songs
top_songs = df1.groupby('track_name')['streams'].sum().sort_values(ascending=False).head(10)

# Create a new DataFrame for plotting
df_plot = df1[df1['track_name'].isin(top_songs.index)]

# Create the interactive bar chart
fig = px.bar(df_plot, x='track_name', y='streams', color='released_year',
             title="Top 10 Streamed Songs with Release Year",
             hover_data=['released_year'],
             labels={'track_name': 'Song Title', 'streams': 'Total Streams', 'released_year': 'Released Year'})

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
value_to_drop = 'Love Grows (Where My Rosemary Goes)'

# Drop the columns
df1 = df1[df1['track_name'] != value_to_drop]
```

```python
top_songs = df1.groupby('track_name')['streams'].sum().sort_values(ascending=False).head(10)


df_plot = df1[df1['track_name'].isin(top_songs.index)]


fig = px.bar(df_plot, x='track_name', y='streams', color='released_year',
             title="Top 10 Streamed Songs with Release Year",
             hover_data=['artist(s)_name','released_year', 'released_month' ],
             labels={'track_name': 'Song Title', 'streams': 'Total Streams', 'released_year': 'Released Year'})

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
top_songs = df1.groupby('artist(s)_name')['streams'].sum().sort_values(ascending=False).head(10)


df_plot = df1[df1['artist(s)_name'].isin(top_songs.index)]


fig = px.bar(df_plot, x='artist(s)_name', y='streams', color='released_year',
             title="Most listened Artist",
             hover_data=['released_year', 'released_month' ],
             labels={'artist(s)_name': 'Artist', 'streams': 'Total Streams', 'released_year': 'Released Year'})

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```
